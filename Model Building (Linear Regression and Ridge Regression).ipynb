#### LoanTenure = (𝑀𝑎𝑡𝑢𝑟𝑖𝑡𝑦𝐷𝑎𝑡e_𝑂𝑟𝑖𝑔𝑖𝑛𝑎𝑙𝑦𝑒𝑎𝑟 − 𝐿𝑜𝑎𝑛𝐷𝑎𝑡𝑒𝑦𝑒𝑎𝑟) 𝑥 12 − (𝑀𝑎𝑡𝑢𝑟𝑖𝑡𝑦𝐷𝑎𝑡𝑒_𝑂𝑟𝑖𝑔𝑖𝑛𝑎𝑙𝑚𝑜𝑛𝑡ℎ − 𝐿𝑜𝑎𝑛𝐷𝑎𝑡𝑒𝑚𝑜𝑛𝑡ℎ)

#### Equated Monthly Installments (EMI):

Tenure ---> **LoanTenure**
Principle repayment ---> **LP_CustomerPrinciplePayments**
Interest ---> **BorrowerRate**

```python
def calculate_emi(LP_CustomerPrincipalPayments,BorrowerRate,LoanTenure):
    monthly_interst_rate = BorrowerRate/12/100
    emi = ((LP_CustomerPrincipalPayments*monthly_interst_rate*(1+monthly_interst_rate)**(LoanTenure))/((1+monthly_interst_rate)**(LoanTenure-1)))
    
    return emi
```

```python
ndf["EMI"]= calculate_emi(ndf["LP_CustomerPrincipalPayments"],ndf["BorrowerRate"],ndf["LoanTenure"])
```

#### Preferred Return on Investment (PROI):

```python
def calculated_roi():
    ndf['InterestAmount']=(ndf['LoanOriginalAmount']*(ndf["BorrowerRate"]))
    ndf["TotalAmount"]=(ndf["InterestAmount"]+ndf["LoanOriginalAmount"])
    ndf["ROI"]=(ndf["InterestAmount"]/ndf["TotalAmount"])
    return ndf["ROI"]
#ndf.drop(['InterestAmount'],axis=1,inplace=True)
#ndf.drop(["TotalAmount"],axis=1,inplace=True)
```

#### Eligible Loan Amount (ELA):

Components of ELA:

A: “AppliedAmount” ---> LoanOriginalAmount
R: “Interest” ---> BorrowerRate
N: “LoanTenure” ---> LoanTenure
I: “IncomeTotal” ---> StatedMonthlyIncome

Calculation Procedure: For each row in the dataset:

Calculate: Total Payment Due = (A + (A*r)) * n
Calculate: Max allowable amount = I * 12 * 30%
If ( Total Payment Due <= Max allowable amount)
  Then ELA = AppliedAmount
  Else ELA = Max allowable amount

```python

def eligible_loan_amount():
    ndf["Avg_Inc"]=(ndf["StatedMonthlyIncome"]*12*0.3)*ndf["LoanTenure"]
    ndf["Total_Loan_Amount"]=np.round(ndf["LoanOriginalAmount"]+(ndf["LoanOriginalAmount"]*ndf["BorrowerRate"])*ndf["LoanTenure"])
    ELA=np.empty(0)
    for i in range(len(ndf["Avg_Inc"].values)):
        if ndf["Avg_Inc"].iloc[i]<=0:
            ELA = np.append(ELA,0)
        elif ndf["Total_Loan_Amount"].iloc[i]<=ndf["Avg_Inc"].iloc[i]:
            ELA = np.append(ELA,ndf["Total_Loan_Amount"].iloc[i])
        else :
            ELA = np.append(ELA,ndf["Avg_Inc"].iloc[i])
    ndf.drop(["Avg_Inc"],axis=1,inplace=True)
    ndf.drop(["Total_Loan_Amount"],axis=1,inplace=True)
    return ELA
            
   
```

```python
ndf["ELA"]=eligible_loan_amount()
ndf["ELA"].describe()
```

## Final Pipeline for Regression Models

```python
from sklearn.ensemble import StackingRegressor
from sklearn.multioutput import MultiOutputRegressor
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
estimators = [('Ridge', Ridge_classifier), ('Linear', LR_model)]
final_estimator = GradientBoostingRegressor()
reg = StackingRegressor(estimators=estimators, final_estimator=final_estimator)
multioutput_reg = MultiOutputRegressor(reg)
```

```python
multioutput_reg
```

```python
model_reg = Pipeline([
    ('scaler', scaler),
    ('regressor',multioutput_reg )
])
```

```python
model_reg.fit(X_train,Y_train)
```

```python
pred = model_reg.predict(X_test)
print('test accuracy = ', round(r2_score(Y_test, pred)*100, 2), '%')
```

```python
mae = round(metrics.mean_absolute_error(Y_test, pred),2)
mse = round(metrics.mean_squared_error(Y_test, pred),2)
print(f" Mean Absolute Error: {mae}\n")
print(f" Mean Squared Error: {mse}\n")
```
