# Prosper Loan Data Project 

## Structure of dataset

#### The Prosper loan dataset comprises of 113937 loan entries with 81 attributes on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others, from the year 2009-2014. There are two main categories:

* Borrower information: Basic attributes of the borrowers such as annual income, condition of employment, interest rate, loan status, etc.
* Loan performance information: Metrics evaluating the risk associated with the loans such as Prosper score and bank card utilization, etc.

As mentioned in the documnetation that the listing category column has some values i.e.
* ListingCategory 
The category of the listing that the borrower selected when posting their listing: 
* 0 - Not Available, 
* 1 - Debt Consolidation,
* 2 - Home Improvement, 
* 3 - Business, 
* 4 - Personal Loan,
* 5 - Student Use,
* 6 - Auto,
* 7- Other,
* 8 - Baby&Adoption, 
* 9 - Boat, 
* 10 - Cosmetic Procedure, 
* 11 - Engagement Ring, 
* 12 - Green Loans,
* 13 - Household E

```python
# convert the ListingCategory (numeric) columns into the ListingCategory column with categories as metioned above 
categ = {0 : 'Not Available', 1 : 'Debt Consolidation', 2 : 'Home Improvement', 3: 'Business', 
             4 : 'Personal Loan', 5 : 'Student Use', 6 : 'Auto', 7 : 'Other', 8 : 'Baby&Adoption', 
             9 : 'Boat', 10 : 'Cosmetic Procedure', 11 : 'Engagement Ring', 12 : 'Green Loans',
             13 : 'Household Expenses', 14 : 'Large Purchases', 15 : 'Medical/Dental', 16 : 'Motorcycle',
             17 : 'RV', 18 : 'Taxes', 19 : 'Vacation', 20 : 'Wedding Loans'}

loan_df['ListingCategory'] = loan_df['ListingCategory (numeric)'].map(categ)
loan_df.drop(['ListingCategory (numeric)'], axis=1, inplace=True)
```

```python
loan_df['ListingCategory'].unique()
```

```python
# convert string to category type of ProsperRating (Alpha) column

values = ['AA', 'A', 'B', 'C', 'D', 'E','HR']
categorical_var = pd.api.types.CategoricalDtype(ordered = True, categories = values)
loan_df['ProsperRating'] = loan_df['ProsperRating (Alpha)'].astype(categorical_var)
```

```python
loan_df.dtypes
```

```python
# Converting the bool values into the numeric such that replacing the "True":1 and "False":0 of columns like ["IsBorrowerHomeowner","IncomeVerifiable"]IsBorrowerHomeowner","IncomeVerifiable"]

#df['col'] = df['col'].map({True: 1, False: 0})  this is the methof that I have used 

loan_df["IncomeVerifiable"]=loan_df["IncomeVerifiable"].map({True: 1, False:0})
# loan_df["IsBorrowerHomeowner"]=loan_df["IsBorrowerHomeowner"].map({True: 1, False:0})
```

```python
loan_df["IsBorrowerHomeowner"]=loan_df["IsBorrowerHomeowner"].map({False: 0, True: 1})
```

```python
# calcute averge Credit Score Range
loan_df['CreditScore_average'] = (loan_df['CreditScoreRangeLower'] + loan_df['CreditScoreRangeUpper'])/2
```

```python
# Create the new binary column 'LoanStatusBinary' based on your criteria
loan_df['LoanStatusBinary'] = loan_df['LoanStatus'].apply(lambda x: 1 if x == 'Completed' else 0)

# Drop the original 'LoanStatus' column
loan_df.drop(columns=['LoanStatus'], inplace=True)

# Rename the new column to 'LoanStatus'
loan_df.rename(columns={'LoanStatusBinary': 'LoanStatus'}, inplace=True)
```

```python
# Find columns with missing values
columns_with_missing_values = loan_df.columns[loan_df.isnull().any()].tolist()

# Display the columns with missing values
print("Columns with missing values:", columns_with_missing_values)
```

```python
# Identify columns with boolean values
boolean_columns = [col for col in loan_df.columns if loan_df[col].dtype == bool]

# Loop through boolean columns and cast as int
for col in boolean_columns:
    loan_df[col] = loan_df[col].astype(int)
```
